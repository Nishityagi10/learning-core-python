### BASICS FOR LOOP IN PYTHON

#### for loop is used to repeat a block of code a specific number of time,or for each item in a sequence(like a list,string,etc)

```python
# example 1
fruits=["apple","banana","cherry"]
for fruit in fruits:
    print(fruit)
```

```python
fruits=["apple","banana","cherry"]
for fruit in fruits:
    print(fruit,end=" ")
```

### using range() with for loop
#### range(start,stop,step) is often used with for to repeat something multiple times

```python
for i in range(1,5,2):
    print(i)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(10,30,4):
    print(i)
```

#### Nested for loop
a nested loop is when one for loop is inside another

```python
#syntax
for loop in outer_loop:
    for j in inner_loop:
        #code block
```

```python
for i in range(3): #outer loop runs 3 times(work in row)
    for j in range(2):  #inner loop runs 2 times for each outer(work in column)
        print(f"i={i},j={j}")
    
```

#### number pyramid

```python
rows=5
for i in range(5,rows-1):
    for j in range(1,i+1):
        print(j, end=" ")
    print()
```

```python
rows=5
for i in range(1,rows+1):
    for j in range(1,i+1):
       
        print(j)
    print()
```

```python
n=6
for i in range(1,n+1):
    for j in range(1,i+1):
        print("*",end="")
    print(j)
```

```python
num=int(input("enter no.of rows and col."))
for i in range(num):
    for j in range(num):
        if i==0 or i==num-1 or j==0 or j==num-1:
            print("*",end="")
        else:
            print(" ",end="")
        print()
```
